## Data Cleaning

Handling Missing Values
- - Drop Missing Values:


```python
df.dropna(axis=0, how='any', subset=['column1', 'column2'])
```

- - Imputation with Mean:



```python
df['column_name'].fillna(df['column_name'].mean(), inplace=True)
```

KNN Imputation:



```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5, weights='uniform')
df_filled = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

Handling Duplicates


```python
df.drop_duplicates(subset=['column1', 'column2'], keep='first')
```

Handling Outliers


```python
Q1 = df['column_name'].quantile(0.25)
Q3 = df['column_name'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['column_name'] < (Q1 - 1.5 * IQR)) | (df['column_name'] > (Q3 + 1.5 * IQR)))]
```

Handling Inconsistent Data


```python
df['date_column'] = pd.to_datetime(df['date_column'], format='%Y-%m-%d', errors='coerce')
```

```python
# Fill missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)

# Drop rows with missing 'rating' or 'date_added'
df.dropna(subset=['rating', 'date_added'], inplace=True)
```

```python
# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'], format='%Y-%m-%d', errors='coerce')
```

```python
df.info()
```

```python
# Check for outliers in 'release_year'
print(df['release_year'].describe())

# Handle outliers (e.g., remove movies released before 1940 or after 2023)
df = df[(df['release_year'] >= 1940) & (df['release_year'] <= 2023)]
```

## Feature Engineering

Feature Engineering

Feature Creation


```python
df['BMI'] = df['weight'] / (df['height'] ** 2)
```

Feature Transformation


```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
df['scaled_column'] = scaler.fit_transform(df[['column_name']])
```

Feature Selection


```python
corr_matrix = df.corr()
high_corr_features = corr_matrix[abs(corr_matrix) > 0.5].index
```

One-Hot Encoding:



```python
df = pd.get_dummies(df, columns=['categorical_column'], drop_first=True)
```

```python
df
```

```python
# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
```

```python
# Extract 'year_added' and 'month_added'
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df
```

```python
df['cast'].str.split(', ')
```

```python
# Create 'num_cast' feature
df['num_cast'] = df['cast'].apply(lambda x: len(x.split(', ')) if x != 'Unknown' else 0)
```

```python
# Perform one-hot encoding
df = pd.get_dummies(df, columns=['type', 'rating'], drop_first=True)
```

```python
# Create 'is_recent' feature
df['is_recent'] = df['year_added'].apply(lambda x: 1 if x > 2020 else 0)
```
