```python
def scrapbyday(date):
    url = f"https://www.yallakora.com/match-center/%D9%85%D8%B1%D9%83%D8%B2-%D8%A7%D9%84%D9%85%D8%A8%D8%A7%D8%B1%D9%8A%D8%A7%D8%AA?date={date}#"
    website = requests.get(url).text
    website = BeautifulSoup(website, "lxml")
    teams = website.find_all("div", class_ = "teamsData")
    for team in teams: 
        teama = team.find("div", class_ = "teams teamA").p.text
        teamb = team.find("div", class_ = "teams teamB").p.text
        print(teama, teamb)
```
