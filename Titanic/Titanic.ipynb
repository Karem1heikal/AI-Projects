```python
data['Age'].fillna(data['Age'].mean(),inplace = True)
data
```

```python
#categorizations
from sklearn.preprocessing import LabelEncoder

Lab_en = LabelEncoder()

df1 = Lab_en.fit_transform(data['Embarked'])
data['Embarked'] = df1
data
```

```python
Lab_en = LabelEncoder()

df1 = Lab_en.fit_transform(data['Sex'])
df1
data['Sex'] = df1
data
```

```python
Lab_en = LabelEncoder()

df1 = Lab_en.fit_transform(data['Ticket'])
df1
data['Ticket'] = df1
data
```

```python
Lab_en = LabelEncoder()

df1 = Lab_en.fit_transform(data['Name'])
df1
data['Name'] = df1
data
```

```python
data['Cabin'].fillna("not_null",inplace = True)
data
```

```python
counts = pd.value_counts(data['Cabin'])
df1 = data['Cabin'].map(counts)
data['Cabin'] = df1
data

# feature Extraction
# if Values of Cabin >3
#else if Values of Cabin == 3 then this is higher probability to be safe
#else if Values of Cabin == 2 then this is Lower probability to be safe than above condition
#else if Values of Cabin == 1   -------------- the same
#else if Values of Cabin == 0   -------------- the same
```
